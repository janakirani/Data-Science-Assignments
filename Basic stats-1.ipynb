**Brief interpretation of these statistics:**
In this Sales & Discount dataset we have 6 numerical columns namely Volume,Avg Price,Total Sales Value,Discount Rate(%),Discount Amount Net Sales Value.As a part of descriptive analytics we found the Mean,Median and Mode to compute basic statistical measures.
Mean,Median and mode are the 3 important measures of central tendency.
Mean=the sum of values divided by number of values.
Median-After arranging the data the middle value(n=1/2 th position) for odd-numberd dataset and for odd-numbered dataset (n/2)+(n/2+1).
Mode-Most occured value in the dataset.
For Normal Distribution Mean=Median=Mode but for skewed distribution Mean,Median,Mode will be different value.Based on the datasert and outliers we have to choose which measure of central tendency is used for the particular dataset.

# **Standardization of Numerical Variables:**

**Standardization:**
We can call it as Z score normalization.Basically this is one of the technique in Feature scaling.

**Feature Scaling:**
If your data has different measurements and different values,It will be very difficult for any one to compare.For examples we can't compare Kg's with Litres or 2 with 200.So in this case scaling plays a major role,by scaling the data we can compare two different values and measurements also.We do have many techniques in Feature Scaling Namely:

1.Absolute Max Scaling 2.min-max Scaling 3.Normalization 4.Standardization

So we can say Standardization is one of those Feature Scaling technique to standardize the independent features of the dartaset.It is performed during the data pre-processing.
The formula for standardization is
z=x-mu/sigma
where x is normal value,mu is mean and sigma is standard deviation.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("/content/discounts.csv")
```

```python
def standardize(column):
    return (column - column.mean()) / column.std()
```

```python
df_standard=df[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']]
```

```python
df_standard= df_standard.apply(standardize)
```

# **Conversion of Categorical Data into Dummy Variables:**

**One-hot encoding:**
Usually when working with real time data applying some machine learning alogorithms is difficult,especially when we're working with Categorical data which are non-numerical.Some ML Algorithms can operate on categorical label data but most of the Algorithms cannot do this.Some Alogorithms can operate only on numeric data for both input and output variables.So,we need to convert the categorical data into numeric form for this we have two techniques
1.Integer Encoding or Label Encoding
2.One-hot Encoding

If the integer values have natural order relationship then Integer Encoding is enough for ML Algorithms to understand.For Example,first/second/third==1,2,3.so,for ordinal variables Integer encoding is best.
If the the values are nominal then we have to use one-hot encoding.we have to use binary variables '0' and '1' in one-hot-encoding.For example,colors green,red,blue,yellow etc...If the color is blue then '1' is placed at blue and '0' at remaining colors and vice versa.We can call these binary variables as dummy variables.
