```python
import pandas as pd
import numpy as np
```

```python
movies_df = pd.read_csv('/content/anime.csv')
```

```python
movies_df.head()
```

```python
#number of unique users in the dataset
len(movies_df.anime_id.unique())
```

```python
user_movies_df = movies_df.pivot(index='anime_id',
                                 columns='name',
                                 values='rating').reset_index(drop=True)
```

```python
user_movies_df
```

```python
user_movies_df.index = movies_df.anime_id.unique()
```

```python
user_movies_df
```

```python
#Impute those NaNs with 0 values
user_movies_df.fillna(0, inplace=True)
```

```python
user_movies_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_movies_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids
user_sim_df.index = movies_df.anime_id.unique()
user_sim_df.columns = movies_df.anime_id.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
